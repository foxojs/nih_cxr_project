```python
import numpy as np
import matplotlib.pyplot as plt

# Data
labels = ["No Finding", "Atelectasis", "Cardiomegaly", "Effusion", "Infiltration",
          "Mass", "Nodule", "Pneumonia", "Pneumothorax", "Consolidation",
          "Edema", "Emphysema", "Fibrosis", "Pleural_Thickening", "Hernia"]

values = [0.565305, 0.244867, 0.187031, 0.432175, 0.396820,
          0.265067, 0.148289, 0.042446, 0.188599, 0.132429,
          0.138874, 0.165934, 0.033422, 0.085493, 0.006697]

# Number of variables
num_vars = len(labels)

# Compute angle for each axis in the plot (in radians)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is made by "closing" the circle, so we need to append the first value at the end
values += values[:1]
angles += angles[:1]

# Create the radar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.plot(angles, values, color='b', linewidth=2, linestyle='solid')
ax.fill(angles, values, color='b', alpha=0.25)

# Add labels to the axes
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10)

# Add a title
ax.set_title("Radar Plot of Best F1 Scores", size=15, y=1.1)

plt.tight_layout()
plt.show()
```
